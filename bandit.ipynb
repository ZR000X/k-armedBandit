```python
import numpy as np
import matplotlib.pyplot as plt

# CODE: is to empirically evaluate expected returns

def ar1(R, Pr, N):
    # R = np.array(R)
    # Pr = np.sort(Pr)
    # for pr in Pr:
    #     assert(pr >= 0 and pr <= 1)
    return np.average([np.random.choice(R, p=Pr) for _ in range(N)])
# def ar2(R, Pr, N):
#     R = np.array(R)
#     Pr = np.sort(Pr)
#     for pr in Pr:
#         assert(pr >= 0 and pr <= 1)
#     return np.average([np.random.choice(R, p=Pr) for _ in range(N)])

def dp1(R, Pr, N):
    x = range(N)
    y = np.array([])
    for j in x:
        a = ar1(R,Pr,j)
        if a > 0:
            y = np.append(y, a)
    fig, ax = plt.subplots()
    ax.set_title(f'Progressive Average Rewards')
    plt.xlabel('N')
    plt.ylabel('Average Reward')
    ax.scatter(x, y, s=0.2)
    plt.axhline(y = -0.39, color = 'r', linestyle = '-')
    print(y)
    print(np.average(y))
    # ax.axhline(y=, color = 'r')
    plt.show()
    print(y[-1])

# def dp2(R, Pr, N):
#     x = range(N)
#     y = np.array([])
#     for j in x:
#         y = np.append(y, ar2(R,Pr,j))
#     fig, ax = plt.subplots()
#     ax.set_title(f'Progressive Average Rewards')
#     plt.xlabel('N')
#     plt.ylabel('Average Reward')
#     plt.axhline(y = -0.39, color = 'r', linestyle = '-')
#     plt.axhline(y = np.average(y), color = 'g', linestyle = '-')
#     ax.scatter(x,y, s=0.2)
#     print(y[-1])

dp1([10,-1,0],[0.01,0.49,0.5],300)
# dp2([10,-1,0],[0.01,0.49,0.5],300)
```
